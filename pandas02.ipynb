```python
#  (1) 대전제1: 데이터프레임에 인덱스([])를 적용 => 열들을 추출

# 한 개 컬럼만 추출할 때 => ['컬럼명']
df_movies['overview']

# 여러 칼럼(리스트)을 추출할때 =>[['컬럼목록리스트']]
df_movies[['budget','original_title']]
```

```python
# (2) 대전제2: 데이터프레임에 슬라이싱([])을 적용 => 행들을 추출 

# df_movies[:] # 모든 행 추출
# df_movies[:3] # 0~2행이 추출
df_movies[100:] # 101행부터 끝까지 추출
# df_movies[49:100] # 50~100행 추출
```

```python
# (3) 대전제3: 데이터프레임에 loc 속성을 이용 => 행들을 추출
#              각 행의 라벨명을 이용해서 행들을 추출

# df_movies.loc[0]  # 0은 행번호가 아니라 행라벨!!
df_movies.loc[0:5]  # loc의 슬라이싱 => ***NOT half-open*** . 하프오픈은 인덱싱할때 해당되는 개념이다. 0, 5는 인덱스가 아니라 라벨명!! 
```

```python
# (4) 대전제4: 데이터프레임에 iloc 속성을 이용 => 행들을 추출

df_movies.iloc[0]

# df_movies.loc[0:]
df_movies.iloc[:]

from pprint import pprint as pp
pp(df_movies.iloc[0:0])

df_movies.iloc[49:100]
```

```python
# (5) 대전제: 부분범위의 행과 열을 동시에 지정해서, Sub-DataFrame을 추출
    
# df_movies[:]['original_language']  # Ok
# df_movies[:][:]   # Ok: 모든행(첫번째 슬라이싱), 모든열(두번째 슬라이싱)

# df_movies[:10][:2]  #  df_movies[:10]으로 데이터 프레임을 만들었고, 거기에 [:2]가 적용되어버림

df_movies[:10][['genres','original_title']]  #행과 열을 함께 지정할때는 열은 반드시 컬럼명을 적는다.
```

```python
# (6) 대전제: loc속성을 이용해서 행과 열의 부분 범위를 지정가능하다


# df_movies.loc[[0,1]][['original_title', 'tagline']]
# df_movies.loc[:,:]  # loc, iloc에서는 슬라이싱 두번 적용가능하다. 
# 첫번째 슬라이싱: 행의 범위 지정
# 두번째 슬라이싱: 열의 범위 지정

df_movies.loc[5:10, 'genres':'keywords']
```
