```python
from pprint import pprint as pp
from inspect import signature as sig
from inspect import getmro
```

```python
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df_tips = sns.load_dataset('tips')

df_tips.head()
```

## seaborn 시각화 라이브러리를 이용한 시각화

### 1. sns.distplot() - Distribution Plot = Histogram + KDE

```python
ax = sns.distplot(df_tips.total_bill)
```

```python
sig(sns.distplot)
```

```python
# sns.distplot(df_tips.total_bill, kde=False)
# sns.distplot(df_tips.total_bill, kde=True, hist=False)
sns.distplot(df_tips.total_bill, kde=True, hist=True, rug=True, bins=30)
```

```python
sig(sns.countplot)
```

```python
df_tips.head()
```

```python
ax = sns.countplot(data=df_tips, x='day')

ax.set_title('Count of each day', fontsize=14)
ax.set_xlabel('Day of Week', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
```

```python
ax = sns.countplot(data=df_tips, x='day', hue='sex')
```

### 3. sns.regplot() - Scatter plot + Regression line

```python
sig(sns.regplot)
```

```python
ax = sns.regplot(data=df_tips, x='total_bill', y='tip')

# ax.set_title(...)
# ax.set_xlabel(...)
# ax.set_ylabel(...)
```

```python
sns.regplot(data=df_tips, x='total_bill', y='tip', lowess=True)
```

```python
sns.regplot(data=df_tips, x='total_bill', y='tip', robust=True)
```

## 4. sns.jointplot() - Scatter plot + Histogram

```python
# ax = sns.jointplot(data=df_tips, x='total_bill', y='tip')
# ax = sns.jointplot(data=df_tips, x='total_bill', y='tip', hue='sex')
# ax = sns.jointplot(data=df_tips, x='total_bill', y='tip', kind='kde')
ax = sns.jointplot(
    data=df_tips,
    x='total_bill',
    y='tip',
    kind='kde',
    hue='sex',
    shade=True
)
```

## 5. sns.violinplot() - Boxplot 을 보완하는 플롯

```python
ax = sns.violinplot(data=df_tips, x='time', y='total_bill')
```

```python
ax = sns.violinplot(data=df_tips, x='time', y='total_bill', hue='sex')
```

## 6. sns.pairplot() - Correlation plot of all variables

```python
# ax = sns.pairplot(data=df_tips)
ax = sns.pairplot(data=df_tips, hue='sex')
```
